# Some Information About Z-Score And Its Table

Since the normal distribution is a probability distribution, then the total area under the curve is 1 or 100%. In standard normal distribution, the area under the curve between two points is calculated and supplied in standard normal table or z-score table or z table. 

Positive scores in the z-score table correspond to the values which are greater than the mean in the following:

```python
import pandas as pd
import numpy as np
import scipy.stats as stats

# We have to import the libraries that are pandas for creating DataFrames, numpy for numerical operations, and scipy.stats for accessing the standard normal distribution function.
# We'll define the index values as Z-scores that ranges from 0 to 4 with a step of 0.1, whose values represent the row indices of the Z-table.
# The column values represent p-values that ranges from 0.00 to 0.09 with a step of 0.01, whose values represent the column indices of the Z-table.
index_values = np.round(np.arange(0, 4.1, 0.1), 2)
columns_values = np.round(np.arange(0.00, 0.1, 0.01), 2)

# We have to initialize an empty list data to store the cumulative probabilities.
data = []

# We'll iterate over each Z-score z_score in the index values and each p-value p_value in the column values.
for z_score in index_values:
    
    row_data = []
    
    for p_value in columns_values:
        
        cumulative_prob = stats.norm.cdf(z_score + p_value)  # The standard normal distribution function
        
        row_data.append(cumulative_prob)  # The cumulative probability cumulative_prob is appended to the row_data list.
        
    data.append(row_data)   # The row_data list is appended to the data list.

z_table = pd.DataFrame(data=data, index=index_values, columns=columns_values) #  creation of the DataFrame z_table by using the data, index, and columns.

print(z_table)
```

Negative scores in the z-table correspond to  the values which are less than the mean in the following:

```python
import pandas as pd
import numpy as np
import scipy.stats as stats

index_values = np.round(np.arange(-4, 0, 0.1), 2)
columns_values = np.round(np.arange(0.00, 0.1, 0.01), 2)

data = []

for z_score in index_values:
    
    row_data = []
    
    for p_value in columns_values:
        
        cumulative_prob = stats.norm.cdf(z_score + p_value)
        
        row_data.append(cumulative_prob)
        
    data.append(row_data)

z_table = pd.DataFrame(data=data, index=index_values, columns=columns_values)

print(z_table)
```

This table allows us to know the percentage of the values usually a decimal figure below to the left of a given z-score on a standard normal distribution. Each z-score correspons to a specific area under the curve. There are two z-score tables. The first one is positive z-score table. It is used when the z-score is positive, and it is above the mean. It enables us to find the percentage or the possibility of all values which happen below a given positive z-score in a standard normal distribution. The second one is negative z-score table. This is used when the z-score is negative, and it is below the mean. It allows us to find the probability of all values that occurs below a given negative z-score in a standard normal distribution. 

This table is very helpful for comparing data points and averages in cases such as test scores, medical measurements and financial investments.

The corresponding values which are greater than the mean are marked with the positive score in the z-table and respresent the area under the bell curve to the left of z.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
# scipy.stats is imported as stats to work with statistical functions.
# We'll create a standard normal distribution with mean = 0 and standard deviation = 1.
mu = 0
sigma = 1

# x is an array of 900 points between mu - 4*sigma and mu + 4*sigma.
# The range from -4 to 4 covers almost the entire area under the standard normal curve.
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 900)
# y is the probability density function of the normal distribution that is evaluated at each point in x.
y = stats.norm.pdf(x, mu, sigma)   

# We'll plot the normal distribution
plt.plot(x, y, label='The Standard Normal Distribution')

# x_fill is an array between mu - 4*sigma and the z-score, which covers the area to the left of the z-score.
x_fill = np.linspace(mu - 4*sigma, z_score)
# y_fill is the probability density function of the normal distribution that is evaluated at each point in x_fill.
y_fill = stats.norm.pdf(x_fill, mu, sigma)

plt.fill_between(x_fill, 0, y_fill, color='purple', alpha = 0.5)
# It fills the area between the curve and the x-axis with the color purple.
# The argument alpha = 0.5 makes the shaded area semi-transparent.

# We have to add a vertical line at the z-score.
plt.axvline(z_score, color='brown', linestyle='--', label=f'z = {z_score}')

# We'll add labels and legend.
plt.title('The Table For Positive Z-Scores In The Standard Normal Distribution')   # setting the title of the plot.
plt.xlabel('X')   # labelling the x-axis
plt.ylabel('The Probability Density')   # labelling the y-axis.

plt.legend()

plt.show()

# For example, let's take z-score = 1.24
z_score = 1.24
```

The corresponding values which are less than the mean are marked with a negative score in the z-table and respresent the area under the bell curve to the left of z.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Creation of the standard normal distribution with mean = 0 and standard deviation = 1
mu = 0
sigma = 1

x = np.linspace(mu - 4*sigma, mu + 4*sigma, 900)
y = stats.norm.pdf(x, mu, sigma)

# We'll plot the normal distribution.
plt.plot(x, y, label='The Standard Normal Distribution')

# We'll shade the area for the negative z-score.
x_fill = np.linspace(mu - 4*sigma, z_score)
y_fill = stats.norm.pdf(x_fill, mu, sigma)

plt.fill_between(x_fill, 0, y_fill, color='grey', alpha=0.5)

# We have to add the vertical line at the z-score.
plt.axvline(z_score, color='orange', linestyle='--', label=f'z = {z_score}')

# We'll add labels and legend.
plt.title('The Table For Negative Z-Scores In The Standard Normal Distribution')

plt.xlabel('X')
plt.ylabel('The Probability Density')

plt.legend()

plt.show()

# For instance, take z-score = -1.07.
z_score = -1.07
```
