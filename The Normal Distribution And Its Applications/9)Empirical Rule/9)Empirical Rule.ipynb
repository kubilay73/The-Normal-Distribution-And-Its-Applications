# Empirical Rule

The Empirical Rule is a statistical rule that states that almost all observed data for a normal distribution will fall within three standard deviations of the mean, which is the average of all of the numbers in the set. According to this rule, about 68% of the data falls within standard deviation σ = 1, about 95% of the data within standard deviation σ = 2 from the mean, and about 99.7% of the data within standard deviation σ = 3 from the mean. In fact, approximately 68% of the data points fall between (μ - σ) and (μ + σ), about 95% of the data points fall between (μ - 2σ) and (μ + 2σ), and about 99.7% of the data points fall between (μ - 3σ) and (μ + 3σ).

Also, the rule is known as the 68-95-99.7 rule. It provides us with the quick estimation of the spread of data and understanding of the probability of certain values that happen in the set of data. 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mu = 0  # the mean
sigma = 1  # the standard deviation

# x is the array of 900 points that ranges from mu - 4*sigma to mu + 4*sigma that covers most of the values under the normal distribution curve.
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 900)
# y is the corresponding values of the probability density function of the normal distribution at points in x.
y = norm.pdf(x, mu, sigma)

# We'll plot 68% interval.
plt.figure(figsize=(9, 6))  # Creation of the new figure with the size of 9 inches by 6 inches.
# We'll plot the normal distribution curve by using the x values and y values with the label 'The Normal Distribution'.
plt.plot(x, y, label='The Normal Distribution')
# We'll fill the area under the curve.
plt.fill_between(x, y, where=((x >= mu - sigma) & (x <= mu + sigma)), color='grey', label='68% Interval')
# We'll add title and labels.
plt.xlabel('x')
plt.ylabel('The Probability Density')
plt.title('The Empirical Rule: 68% Interval')

plt.legend()
plt.grid(True)
plt.show()

# We'll plot 95% interval.
plt.figure(figsize=(9, 6))
plt.plot(x, y, label='The Normal Distribution')
# Filling the area under the curve
plt.fill_between(x, y, where=((x >= mu - 2*sigma) & (x <= mu + 2*sigma)), color='pink', label='95% Interval')
# We'll add title and labels.
plt.xlabel('x')
plt.ylabel('The Probability Density')
plt.title('The Empirical Rule: 95% Interval')

plt.legend()
plt.grid(True)
plt.show()

# We'll plot 99.7% interval.
plt.figure(figsize=(9, 6))
plt.plot(x, y, label='The Normal Distribution')
# Filling the area under the curve
plt.fill_between(x, y, where=((x >= mu - 3*sigma) & (x <= mu + 3*sigma)), color='green', label='99.7% Interval')
# We'll add title and labels.
plt.xlabel('x')
plt.ylabel('The Probability Density')
plt.title('The Empirical Rule: 99.7% Interval')

plt.legend()
plt.grid(True)
plt.show()
```
