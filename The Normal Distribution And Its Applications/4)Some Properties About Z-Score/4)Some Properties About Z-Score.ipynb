# Some Properties About Z-Score

There are some ways of finding normal probabilities in z-table.

1)To find the probability that z is greater than 0, we have to look up the z-score and substract 0.5.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Calculation of the probability P(0 < z < z0) for a standard normal distribution, where z0 is the upper bound z-score.
def probability_between_0_and_z0(z0):
    
# stats.norm.cdf(z0) calculates the cumulative probability up to z0.
# stats.norm.cdf(0) calculates the cumulative probability up to 0.
    prob_between_0_and_z0 = stats.norm.cdf(z0) - stats.norm.cdf(0)
    return prob_between_0_and_z0

def plot_probability_between_0_and_z0(z0):

    x = np.linspace(-4, 4, 900)   # creating the array of 900 points between -4 and 4
    y = stats.norm.pdf(x)  # Computing the probability density function values for a standard normal distribution at each value in x, whose result is an array y. 

    fig, ax = plt.subplots(figsize=(9, 6))  # Creating a figure fig and the set of subplots ax.
# The figsize parameter specifies the size of the figure in inches.
# We'll plot the standard normal distribution.
    ax.plot(x, y, label='The Standard Normal Distribution')

# We have to fill the area for P(0 < Z < z0).
    x_fill = np.linspace(0, z0)   # Creating an array x_fill ranging from 0 to z0.
    y_fill = stats.norm.pdf(x_fill)  # Computing the values of the probability density function for the standard normal distribution at each value in x_fill whose result is an array y_fill
    
    ax.fill_between(x_fill, 0, y_fill, color='grey', alpha=0.5, label=f'P(0 < Z < {z0})')
# It fills the area under the standard normal curve between x_fill and the x-axis with the grey color. 
# The parameter alpha sets the transparency level of the filled area.
# We'll add labels and title.
    ax.set_xlabel('Z-Score')  # Setting the label for the x-axis to "Z-Score".
    ax.set_ylabel('The Probability Density')   # Setting the label for the y-axis to "The Probability Density".
    
    ax.set_title('The Standard Normal Distribution With The Probability')   # Setting the title of the plot.
    
    ax.legend()

    plt.show()

# For example, take z0 = 1.34
z0 = 1.34
prob_between_0_and_z0 = probability_between_0_and_z0(z0)

print(f"P(0 < Z < {z0}) = {prob_between_0_and_z0:.4f}")

plot_probability_between_0_and_z0(z0)   # Plotting the probability
```

2)The symmetry of the normal distribution helps us to calculate probabilities for negative values of z by using the table for the following. In addition, P(-a<z<0) = P(0<z<a).

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Computation of the probability P(-a < Z < 0) for a standard normal distribution.
def probability_between_negative_a_and_0(a):
    
    prob_between_negative_a_and_0 = stats.norm.cdf(0) - stats.norm.cdf(-a)
    return prob_between_negative_a_and_0

# Calculation of the probability P(0 < Z < a) for a standard normal distribution.
def probability_between_0_and_a(a):
    
    prob_between_0_and_a = stats.norm.cdf(a) - stats.norm.cdf(0)
    return prob_between_0_and_a

# We'll plot the probabilities P(-a < Z < 0) and P(0 < Z < a) on a standard normal distribution graph.
def plot_probabilities(a):

    x = np.linspace(-4, 4, 900)
    y = stats.norm.pdf(x)

    fig, ax = plt.subplots(figsize=(9, 6))

# We have to plot the standard normal distribution.
    ax.plot(x, y, label='The Standard Normal Distribution')

# We'll fill the area for P(-a < Z < 0).
    x_fill_negative = np.linspace(-a, 0)
    y_fill_negative = stats.norm.pdf(x_fill_negative)
    
    ax.fill_between(x_fill_negative, 0, y_fill_negative, color='orange', alpha=0.5, label=f'P(-{a} < Z < 0)')

# We'll fill the area for P(0 < Z < a).
    x_fill_positive = np.linspace(0, a)
    y_fill_positive = stats.norm.pdf(x_fill_positive)
    
    ax.fill_between(x_fill_positive, 0, y_fill_positive, color='purple', alpha=0.5, label=f'P(0 < Z < {a})')

# We'll add labels and title.
    ax.set_xlabel('Z-Score')
    ax.set_ylabel('The Probability Density')
    
    ax.set_title('The Standard Normal Distribution With The Probabilities')
    
    ax.legend()

    plt.show()

# For example, take a = 1.16.
a = 1.16

prob_negative_a_to_0 = probability_between_negative_a_and_0(a)
prob_0_to_a = probability_between_0_and_a(a)

print(f"P(-{a} < Z < 0) = {prob_negative_a_to_0:.4f}")
print(f"P(0 < Z < {a}) = {prob_0_to_a:.4f}")

plot_probabilities(a)   # Plotting the probabilities
```

3)P(-b<z<c) = P(-b<z<0) + P(0<z<c) = P(0<z<b) + P(0<z<c) as we use the property of the symmetry of the normal distribution.

```python
import scipy.stats as stats

# Calculation of the probability P(-b < Z < c) for a normal distribution
def probability_between_negative_b_and_c(b, c):
    
    prob_between_negative_b_and_c = stats.norm.cdf(c) - stats.norm.cdf(-b)
    return prob_between_negative_b_and_c

# We'll plot the probability P(-b < Z < c) on a standard normal distribution graph.
def plot_probability_between_neg_b_and_c(b, c):

    x = np.linspace(-4, 4, 900)
    y = stats.norm.pdf(x)

    fig, ax = plt.subplots(figsize=(9, 6))

# We'll plot the standard normal distribution
    ax.plot(x, y, label='The Standard Normal Distribution')

# We have to fill the area for P(-b < Z < c)
    x_fill = np.linspace(-b, c)  # Generating points between -b and c.
    
    y_fill = stats.norm.pdf(x_fill)  # Computing the values of the probability density function for these points.
    
    ax.fill_between(x_fill, 0, y_fill, color='grey', alpha=0.5, label=f'P(-{b} < Z < {c})')

# We'll add labels and title.
    ax.set_xlabel('Z-Score')
    ax.set_ylabel('The Probability Density')
    
    ax.set_title('The Standard Normal Distribution With The Probability')
    
    ax.legend()

    plt.show()

# For example, take b = 1.2 and c = 2.4.
b = 1.2
c = 2.4

prob_negative_b_to_c = probability_between_negative_b_and_c(b, c)

print(f"P(-{b} < Z < {c}) = {prob_negative_b_to_c:.4f}")

plot_probability_between_neg_b_and_c(b, c)   # Plotting the probability
```

4)P(z<-b) = 1 - P(z>-b) = 1 - P(z<b) by using the symmetry of the normal distribution table.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Calculation of the probability P(Z < -b) for a standard normal distribution
def probability_less_than_minus_b(b):

    prob_less_than_minus_b = stats.norm.cdf(-b)
    
    return prob_less_than_minus_b

# We'll plot the probability P(Z < -b) on the standard normal distribution graph.
def plot_probability_less_than_minus_b(b):

    x = np.linspace(-4, 4, 900)  # Creating the array of 900 points between -4 and 4
    y = stats.norm.pdf(x)  # Computation of the corresponding y-values y by using the probability density function of the standard normal distribution stats.norm.pdf(x).

    fig, ax = plt.subplots(figsize=(9, 6))

# We'll plot the standard normal distribution.
    ax.plot(x, y, label='The Standard Normal Distribution')

# We have to fill the area for P(Z < -b).
    x_fill = np.linspace(-4, -b)
    y_fill = stats.norm.pdf(x_fill)
    
    ax.fill_between(x_fill, 0, y_fill, color='red', alpha = 0.5, label=f'P(Z < -{b})')
# It helps us to identify the area under the curve for P(Z < b) with a red color and set the transparency to 0.5. 
   
    ax.set_xlabel('Z-Score')   
    ax.set_ylabel('The Probability Density')   
    ax.set_title('The Standard Normal Distribution With The Probability')
    
    ax.legend()

    plt.show()

# For instance, take b = 1.7
b = 1.7

prob_less_than_minus_b = probability_less_than_minus_b(b)
print(f"P(Z < -{b}) = {prob_less_than_minus_b:.4f}")

plot_probability_less_than_minus_b(b)  # Plotting the probability
```

5)P(z<a) = 0.5 + P(0<z<a), where a is greater than zero.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Calculation of the probability P(Z < a) for standard normal distribution, where a is greater than zero.
def probability_less_than_a(a):

    prob_less_than_a = stats.norm.cdf(a)
    return prob_less_than_a

# We have to plot the probability P(Z < a) on a standard normal distribution graph.
def plot_probability_less_than_a(a):

    x = np.linspace(-4, 4, 900)
    y = stats.norm.pdf(x)

    fig, ax = plt.subplots(figsize=(9, 6))

# We can plot the standard normal distribution.
    ax.plot(x, y, label='The Standard Normal Distribution')

# We have to fill the area for P(Z < a)
    x_fill = np.linspace(-4, a)
    y_fill = stats.norm.pdf(x_fill)
    
    ax.fill_between(x_fill, 0, y_fill, color='green', alpha=0.5, label=f'P(Z < {a})')

# Add labels and title.
    ax.set_xlabel('Z-Score')
    ax.set_ylabel('The Probability Density')
    
    ax.set_title('The Standard Normal Distribution With The Probability')
    
    ax.legend()

    plt.show()

# For example, take a = 2.1.
a = 2.1
prob_less_than_a = probability_less_than_a(a)

print(f"P(Z < {a}) = {prob_less_than_a:.4f}")

plot_probability_less_than_a(a)  # Plotting the probability
```

6)P(e<z<f) = P(0<z<f) - P(0<z<e), where e and f are greater than zero.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# We'll calculate the probability P(e < Z < f) for a standard normal distribution, where e and f are greater than zero.
# Also, e is the lower bound z-score and f is the upper bound z-score.
def probability_between_e_and_f(e, f):

    prob_between_e_and_f = stats.norm.cdf(f) - stats.norm.cdf(e)
    return prob_between_e_and_f

# We have to plot the probability P(e < Z < f) on a standard normal distribution.
def plot_probability_between_e_and_f(e, f):

    x = np.linspace(-4, 4, 900)
    y = stats.norm.pdf(x)

    fig, ax = plt.subplots(figsize=(9, 6))

# We'll plot the standard normal distribution.
    ax.plot(x, y, label='The Standard Normal Distribution')

# We have to fill the area for P(e < Z < f).
    x_fill = np.linspace(e, f, 1000)
    y_fill = stats.norm.pdf(x_fill)
    
    ax.fill_between(x_fill, 0, y_fill, color='brown', alpha=0.5, label=f'P({e} < Z < {f})')

# We have to add labels and title.
    ax.set_xlabel('Z-Score')
    ax.set_ylabel('The Probability Density')
    ax.set_title('The Standard Normal Distribution With The Probability')
    ax.legend()

    plt.show()

# For example, take e = 0.8 and f = 1.9
e = 0.8
f = 1.9

prob_between_e_and_f = probability_between_e_and_f(e, f)
print(f"P({e} < Z < {f}) = {prob_between_e_and_f:.4f}")

plot_probability_between_e_and_f(e, f)   # Plotting the probability
```

7)P(z>g) = 0.5 - P(0<z<g), where g is greater than zero.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Calculation of the probability P(Z > g) for a standard normal distribution, where g is greater than zero.
def probability_greater_than_g(g):

    prob_greater_than_g = stats.norm.sf(g)  # It calculates the probability P(Z>g) by using the survival function stats.norm.sf(g).
    return prob_greater_than_g

# We'll plot the probability P(Z > g) on a standard normal distribution graph.
def plot_probability_greater_than_g(g):

    x = np.linspace(-4, 4, 900)
    y = stats.norm.pdf(x)

    fig, ax = plt.subplots(figsize=(9, 6))

# We have to plot the standard normal distribution.
    ax.plot(x, y, label='The Standard Normal Distribution')

# We'll fill the area for P(Z > g).
    x_fill = np.linspace(g, 4)
    y_fill = stats.norm.pdf(x_fill)
    
    ax.fill_between(x_fill, 0, y_fill, color='black', alpha=0.5, label=f'P(Z > {g})')

# We'll add labels and title.
    ax.set_xlabel('Z-Score')
    ax.set_ylabel('The Probability Density')
    
    ax.set_title('The Standard Normal Distribution With The Probability')
    
    ax.legend()

    plt.show()

# For instance, take g = 1.1.
g = 1.1
prob_greater_than_g = probability_greater_than_g(g)

print(f"P(Z > {g}) = {prob_greater_than_g:.4f}")

plot_probability_greater_than_g(g)   # Plotting the probability
```
