# The Introduction Of The Normal Distribution

  A normal distribution is a type of continuous probability distribution for a real-valued random variable. It is also symmetric about the mean. This shows that values near the mean are more frequent in occurrence than the values that are far from the mean. It is more popular and better measure than other probability distributions such as Binomial, Geometric, Poisson, Uniform and Exponential. Graphically, a normal distribution appears as a bell curve due to its flared shape. The precise shape can vary according to the distribution of the values in the population. Then, in a normal distribution bell curve, the peak is always in the middle. Furthermore, the mean, the mode and the median are all the same. 
  
  The general formula of its probability density function is $
f(x) = \frac{1}{\sigma\sqrt{2\pi}} e^{-\frac{(x - \mu)^2}{2\sigma^2}}
$, where $\pi = \frac{22}{7} = 3.14159...$, e = 2.71828..., $-\infty < x < \infty$. Let's look at the example in the following:

```python
import math
# Calculation of the probability density function of the normal distribution
def normal_pdf(x, mu, sigma):
# x is the point at which we can evaluate the probability density function.
# mu is the mean of the distribution.
# sigma is the standard deviation of the distribution.
    coefficient = 1/(sigma * math.sqrt(2 * math.pi))
    exponent = (-1/2) * (((x - mu)/sigma) ** 2)
    
    f = coefficient * math.exp(exponent)
    return f
```

```python
x = 2
mu = 0
sigma = 1

pdf_value = normal_pdf(x, mu, sigma)

print(pdf_value)
```

Also, in this formula, the parameter μ is the mean or expectation of the distribution, a parameter σ is its standard deviation, x is the value of the parameter and f(x) is the probability function. Then, the variance of the distribution is $σ^2$.

We'll show that why the normal distribution is symmetric.

```python
# We'll show that why the normal distribution is symmetric.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
# We calculate the pdf values for the x values by using scipy.stats.norm.pdf.

mu = 0  # the mean
sigma = 1  # the standard deviation

# We'll generate the range of x values around the mean from mu - 2 * sigma to mu + 2 * sigma).
x = np.linspace(mu - 2*sigma, mu + 2*sigma)

# Calculation of the values of the probability density function for the normal distribution
pdf_values = norm.pdf(x, mu, sigma)

# We'll plot the normal distribution.
plt.plot(x, pdf_values, label='Normal Distribution PDF')

# We can highlight the mean and one standard deviation away from the mean on both sides.
plt.axvline(mu, color='g', linestyle='--', label='μ')
plt.axvline(mu - sigma, color='r', linestyle='--', label='μ - σ')
plt.axvline(mu + sigma, color='r', linestyle='--', label='μ + σ')

# The symmetric points
x1 = mu - 1
x2 = mu + 1
y1 = norm.pdf(x1, mu, sigma)
y2 = norm.pdf(x2, mu, sigma)

# We have to mark two symmetric points around the mean.
plt.plot([x1, x2], [y1, y2], 'yo', label='The Symmetric Points')

plt.title('The Normal Distribution (μ = 0, σ = 1)')
plt.xlabel('x')
plt.ylabel('The Probability Density Function Values')

plt.legend()
plt.grid(True)

plt.show()
```

Let's look at the effect of standard deviation on the probability density function of the normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
# We'll calculate the pdf values for the standard deviations by using scipy.stats.norm.pdf.

mu = 0  # the mean of the distributions
sigmas = [1, 2, 3]  # different standard deviations

# We have to generate the range of x values of 900 points around the mean, which ensures that the range is wide enough to cover all the distributions.
x = np.linspace(mu - 3 * max(sigmas), mu + 3 * max(sigmas), 900)

# We'll plot the normal distributions with different standard deviations
plt.figure(figsize=(9, 6))

for sigma in sigmas:
    
    pdf_values = norm.pdf(x, mu, sigma)
    plt.plot(x, pdf_values, label = f'μ = {mu}, σ = {sigma}')

plt.title('The Effect Of Standard Deviation On The Probability Density Function Of The Normal Distribution')
plt.xlabel('x')
plt.ylabel('The Probability Density Function Values')

plt.legend()
plt.grid(True)

plt.show()
```

Let's look at the effect of mean on probability density function of the normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
# We have to calculate the values of the pdf for the means by using scipy.stats.norm.pdf.

sigma = 1  # the standard deviation of the distributions
means = [-1, 0, 1]  # the means are different here.

# We'll generate the range of x values of 900 points around the smallest and largest means.
x = np.linspace(min(means) - 4 * sigma, max(means) + 4 * sigma, 900)

# We have to plot the normal distributions with different means.
plt.figure(figsize=(9, 6))

for mu in means:
    pdf_values = norm.pdf(x, mu, sigma)
    plt.plot(x, pdf_values, label=f'μ = {mu}, σ = {sigma}')

plt.title('The Effect Of Mean On Probability Density Function Of The Normal Distribution')
plt.xlabel('x')
plt.ylabel('The Probability Density Function Values')

plt.legend()
plt.grid(True)

plt.show()
```

Standard normal distribution is the special normal distribution when the mean μ is 0 and the standard deviation σ is 1. In addition, it is called the z-distribution or unit normal distribution. It is helpful and useful for us to calculate the probability of certain values, which happens in our distribution easily and simply. In fact, it enables us to compare the sets of data with different means and standard deviations with ease. The normal distribution is referred to as N(μ, $σ^2$). Then, if a random variable X is normally distributed with μ and σ , we can write X $\sim N(μ, σ^2)$, where “~” means “distributed as”. 

In standard normal distribution, we have z or z-score. It is a statistical measure that describes the relationship of the value to the mean of the group of values. That is, it quantifies the distance between a data point and the mean of a data set. z-score is measured with respect to the standard deviations from the mean. In standard normal distribution, we convert the individual value into a z-score. The formula for the z-score of the value is given by $
z = \frac{x - \mu}{\sigma}
$, where x is the data point, $\mu$ is the mean of the set of the data, and $\sigma$ is the standard deviation of the dataset.

For this formula, if X is greater than the mean, then z score is positive. Also, if X is equal to the mean, then, z score is zero. In standard normal distribution, if X $\sim N(μ, σ^2)$, then we have z $\sim N(0,1)$, the normal distribution with the mean 0 and standard deviation 1.

```python
def z_score(x, mu, sigma):

    z = (x - mu) / sigma
    return z

x = 5
mu = 0
sigma = 1

z = z_score(x, mu, sigma)

print(f"The z-score of x is {z}")
```
