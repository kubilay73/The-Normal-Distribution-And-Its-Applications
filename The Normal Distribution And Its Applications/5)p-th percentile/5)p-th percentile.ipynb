# p-th percentile

The p-th percentile of the data set of measurements is the value for which at most p% of them are less than that value, and at most (100-p)% of them are above it after the values of the set are ordered from smallest to largest in the standard normal distribution.

To find the p-th percentile in the normal distribution, we can use the inverse of the cumulative distribution function. Firstly, we have to determine the z-score. The z-score that corresponds to the p-th percentile is found by using the inverse of the cumulative distribution function in the standard normal distribution, which is denoted by $Z_p$. The z-score tells us how many standard deviations away from the mean the p-th percentile value is. Secondly, we will convert the z-score to the percentile value. In this case, if μ and σ are the mean and standard deviation of the normal distribution, respectively, then the p-th percentile value $X_p$ can be found by the formula: $$X_p = \mu + Z_p \cdot \sigma$$

There is an example in the following:

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# We'll calculate the pth percentile in the standard normal distribution and plot it.
def plot_pth_percentile(p):

# p is the percentile between 0 and 100.
# We have to convert the percentile to the proportion.
    proportion = p / 100
    
# Calculation of the z-score for the pth percentile.
    z_value = stats.norm.ppf(proportion)
    
# We can define the range for the Z-scores of 900 points from -4 to 4, and calculate the probability density function values for the Z-scores.
    x = np.linspace(-4, 4, 900)
    y = stats.norm.pdf(x)
    
    fig, ax = plt.subplots(figsize=(9, 6))
    
# We'll plot the standard normal distribution
    ax.plot(x, y, label='The Standard Normal Distribution')
    
# We'll highlight the area up to the pth percentile.
    x_fill = np.linspace(-4, z_value)
    y_fill = stats.norm.pdf(x_fill)
    
    ax.fill_between(x_fill, 0, y_fill, color='purple', alpha=0.5, label=f'The {p}th Percentile (Z = {z_value:.2f})')
    
# We have to add the vertical line at the pth percentile
    ax.axvline(z_value, color='orange', linestyle='--')
    
# We can add labels and title.
    ax.set_xlabel('Z-Score')
    ax.set_ylabel('The Probability Density')
    
    ax.set_title(f'The {p}th Percentile In The Standard Normal Distribution')
    ax.legend()
    
    
    plt.show()

# For example, take p = 68.
p = 68  # 68th percentile
plot_pth_percentile(p)
```

In the standard normal distribution, the mean is the 50th percentile, which is typically denoted as Q2, because 50% of the values of the data set fall at or below it.

Also, Q1 is commonly and generally called the lower quartile, and it is 25th percentile. Q3 is often called the upper quartile, and it is 75th percentile.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# We'll calculate the 25th, 50th and 75th percentiles in the standard normal distribution and plot them.
def plot_percentile(p):

# p is the percentile between 0 and 100.    
# We can convert the percentile to the proportion.
    proportion = p / 100
    
# Calculation of the z-score for the pth percentile.
    z_value = stats.norm.ppf(proportion)
    
# We have to define the range for the z-scores from -4 to 4 in 900 points, and calculate the probability density function values for them.
    x = np.linspace(-4, 4, 900)
    y = stats.norm.pdf(x)
    
    fig, ax = plt.subplots(figsize=(9, 6))
    
# The standard normal distribution is plotted.
    ax.plot(x, y, label='The Standard Normal Distribution')
    
# We'll highlight the area up to the pth percentile
    x_fill = np.linspace(-4, z_value)
    y_fill = stats.norm.pdf(x_fill)
    
    ax.fill_between(x_fill, 0, y_fill, color='purple', alpha=0.5, label=f'The {p}th Percentile (Z = {z_value:.2f})')
    
# We can add the vertical line at the pth percentile.
    ax.axvline(z_value, color='pink', linestyle='--')
    
# We can add labels and title.
    ax.set_xlabel('Z-Score')
    ax.set_ylabel('The Probability Density')
    
    ax.set_title(f'The {p}th Percentile In The Standard Normal Distribution')
    
    ax.legend()
    
    plt.show()


percentiles = [25, 50, 75]

for p in percentiles:
      
    plot_percentile(p)  
```

If there are n observations, which are arranged in the increasing order, then the first quartile is at position $\frac{n+1}{4} $, second quartile (i.e. the mean) is at position $\frac{2(n+1)}{4}$, and the third quartile is at position $\frac{3(n+1)}{4}$ in the standard normal distribution.
