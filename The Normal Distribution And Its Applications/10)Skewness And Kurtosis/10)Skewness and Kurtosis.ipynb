# Skewness

It measures the degree of symmetry of a distribution. The normal distribution is symmetric and it has a skewness of zero. The skewness formula is given by $$\text{Skewness} = \frac{{\sum_{i=1}^{n}(x_i - \bar{x})^3}}{{n(\sum_{i=1}^{n} (x_i - \bar{x})^2)^\frac{3}{2}}}$$, where n is the number of observation in the sample, $\bar{x}$ is the sample mean, and $x_i$ represents the observations of the individuals.

If the distribution is negative skewness, the skewness less than zero, then it is left-skewness, and the left tail of the distribution is longer than the right tail.

If the distribution is positive skewness, the skewness greater than zero, then it is right-skewness, and the right tail of the distribution is longer than the left.

As a consequence, positively skewed distributions and negatively skewed distributions are not normal distributions.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma

shape = 3   # the shape parameter for the gamma distribution
scale = 2   # the scale parameter for the gamma distribution
sample_size = 873  # the sample size 

# Generation of the data from a gamma distribution by using np.random.gamma
data_positively_skewed = np.random.gamma(shape, scale, sample_size)

x = np.linspace(0, 20, 873)  # The range of x values

# Calculation of the probability density function of the positively skewed gamma distribution
positively_skewed_curve = gamma.pdf(x, shape, scale=scale)  

# We'll plot the positively skewed gamma distribution curve.
plt.figure(figsize=(9, 6))   # The new figure with the size of 9 inches in width and 6 inches in height
plt.plot(x, positively_skewed_curve, label='The Positively Skewed Gamma Distribution', color='black')

# We'll add the title and the labels.
plt.title('The Positively Skewed Gamma Distribution')
plt.xlabel('The Values')
plt.ylabel('The Probability Density')

plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma

# The parameters for the gamma distribution
shape = 2  # the shape parameter for the gamma distribution
scale = 1  # the scale parameter for the gamma distribution
sample_size = 841  # the sample size parameter for the gamma distribution  

# Generation of the data from the gamma distribution
x = np.linspace(0, 10, 841)  # The range of x values in the gamma distribution
gamma_curve = gamma.pdf(x, shape, scale=scale)  # the probability density function of the gamma distribution

# We have to create the negatively skewed distribution by reflecting the gamma distribution.
# x[::-1] reverses the order of the x values. This mirrors the distribution around the y-axis.
# -x[::-1] + 10 shifts the mirrored x values to the right. The value 10 is chosen to ensure the resulting values are positive.
x_negatively_skewed = -x[::-1] + 10  
negatively_skewed_curve = gamma.pdf(x[::-1], shape, scale=scale)  # the probability density function of the reflected distribution

# We'll plot the negative skewed gamma distribution.
plt.figure(figsize=(9, 6))  # the new figure with the size of 9 inches in width and 6 inches in height
plt.plot(x_negatively_skewed, negatively_skewed_curve, label='Negatively Skewed Gamma Distribution', color='green')

# We'll add the title and the labels.
plt.title('The Negatively Skewed Gamma Distribution')
plt.xlabel('The Values')
plt.ylabel('The Probability Density')

plt.legend()
plt.grid(True)

plt.show()
```

# Kurtosis

It is the measure of the combined weight of the tails of the distribution, which is relative to the center of the curve of the distribution. The normal distribution has a kurtosis equal to 3. Distributions with kurtosis greater than 3 show tail data that exceeds the tails of the normal distribution. The excess kurtosis is known in statistics as leptokurtic. It is also known as "fat tails". The occurrence of it in financial markets describes what is known as fat risk.

Distributions with low kurtosis less than 3, which is platykurtic, show tails that are generally less extreme, or skinner, than the tails of the normal distribution.

Mesokurtic distributions have a kurtosis of 3. They have the same kurtosis as that of the normal distribution or normal curve, also known as a bell curve.

The Kurtosis formula is given by:

$$\text{Kurtosis} = \frac{{\frac{1}{n} \sum_{i=1}^{n} (x_i - \bar{x})^4}}{{\left(\frac{1}{n} \sum_{i=1}^{n} (x_i - \bar{x})^2\right)^2}}$$

, where n is the number of the observations in the sample, $\bar{x}$ is the sample mean, and $x_i$ represents the observations of the individuals.

It calculates the ratio of the fourth moment to the square of the second moment, which supplies the measure of the "tailedness" that is relative to the dispersion of the data. Kurtosis is generally described with respect to excess kurtosis, which is "kurtosis − 3". Because normal distributions have a kurtosis of 3, normal distributions have an excess kurtosis of 0. As a result, any distribution with an excess kurtosis of 0 is mesokurtic.

In addition, since platykurtic distributions have less kurtosis than the normal distribution, then they have an excess kurtosis of less than 0. On the other hand, because leptokurtic distributions have more kurtosis than the normal distribution, then they have an excess kurtosis of more than 0.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# The parameters for the normal distribution
mu = 0  # the mean
sigma = 1   # the standard deviation
sample_size = 950

# We'll generate the data for mesokurtic distribution by using np.random.normal.
mesokurtic_data = np.random.normal(mu, sigma, sample_size)

# We'll generate the data for platykurtic distribution by using np.random.normal by using the smaller standard deviation for the same mean and sample size.
platykurtic_data = np.random.normal(mu, sigma/2, sample_size)

# We'll generate the data for leptokurtic distribution by concatenating two normal distributions.
# One shifted to the left and one shifted to the right of the mean in concatenating them.
leptokurtic_data = np.concatenate([np.random.normal(mu - sigma, sigma, sample_size // 2), np.random.normal(mu + sigma, sigma, sample_size // 2)])

x = np.linspace(-4, 4, 900)  # The range of x values
normal_curve = norm.pdf(x, mu, sigma)  # The probability density function of the normal distribution

plt.figure(figsize=(9, 6))  # The new figure with the size of 9 inches in width and 6 inches in height.

# We'll plot the mesokurtic distribution.
plt.plot(x, normal_curve, label='The Mesokurtic Distribution', color='red')

# The platykurtic distribution is plotted.
platykurtic_mu = np.mean(platykurtic_data)   # the mean of the platykurtic distribution data
platykurtic_sigma = np.std(platykurtic_data)  # the standard deviation of the platykurtic distribution data
platykurtic_curve = norm.pdf(x, platykurtic_mu, platykurtic_sigma)   # computing the probability density function values for the platykurtic distribution by using the mean and the standard deviation.
# We'll plot the probability density function of the platykurtic distribution and label it as 'The Platykurtic Distribution' with the color black.
plt.plot(x, platykurtic_curve, label='The Platykurtic Distribution', color='black')

# We have to plot the leptokurtic distribution.
leptokurtic_mu = np.mean(leptokurtic_data)   # the mean of the leptokurtic data
leptokurtic_sigma = np.std(leptokurtic_data)   # the standard deviation of the leptokurtic data
leptokurtic_curve = norm.pdf(x, leptokurtic_mu, leptokurtic_sigma)   # computing the probability density function values for the leptokurtic distribution by using the mean and the standard deviation.
# Plotting the probability density function of the platykurtic distribution and label it as 'The Platykurtic Distribution' with the color green.
plt.plot(x, leptokurtic_curve, label='The Leptokurtic Distribution', color='green')

# The title and labels should be added.
plt.title('The Types Of Kurtosis')
plt.xlabel('The Values')
plt.ylabel('The Probability Density')

plt.legend()
plt.grid(True)

plt.show()
```
