# The Hypothesis Testing

It is used to estimate the relationship between two statistical variables. Also, it is a type of statistics, where we put our assumptions about a parameter to the test.

In hypothesis testing, the null hypothesis $H_0$ states that the population is normally distributed, against the alternative hypothesis $H_a$, which is not normally distributed. If the test p-value is less than the significance level, then we can reject the null hypothesis and conclude a data that is not from a population with the normal distribution. If the p-value is greater than significance level $\alpha\$, then we do not reject the null hypothesis.

The formula of the hypothesis testing is given by $$z = \frac{\bar{X} - \mu_0}{\frac{\sigma}{\sqrt{n}}}$$, where $\sigma$ is the standard deviation, n is the sample size, $\mu_0$ is the population mean, and $\bar{X}$ is the sample mean. It is valid for one sample.

For two samples, the formula is given by $$z = \frac{(\bar{x}_1 - \bar{x}_2) - (\mu_1 - \mu_2)}{\sqrt{\frac{\sigma_1^2}{n_1} + \frac{\sigma_2^2}{n_2}}}$$, where $n_1$ and $n_2$ are  the sample sizes, $\mu_1$ and $\mu_2$ are the population means, and $\sigma_1^2$ and $\sigma_2^2$ are the sample variances.

p-value is between 0 and 1. It is compared to $\alpha$ or significance level. The $\alpha$ level can be defined as the acceptable risk of incorrectly rejecting the null hypothesis. The $\alpha$ level is usually chosen between 1% and 5%.

When we use the standard normal distribution or z distribution, the test statistic is the standardized value which is the boundary of the p-value. We have known that $z = \frac{x - \bar{x}}{s}$ is the formula for the z score.  The formula for a test statistic will be similar. When we do the hypothesis test, the sampling distribution is centered on the null parameter, and the standard deviation is known as the standard error.

The general form of the test statistic is $$\text{test statistic} = \frac{\text{sample statistic} - \text{null parameter}}{\text{standard error}}$$

This formula puts the observed sample statistic on the standard scale. The z-score tells us where the score lies on the normal distribution in standard deviation. The test statistic tells us where the sample statistic falls on the sampling distribution in the standard error.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Setting the random number generator to the fixed state so that the results are reproducible
np.random.seed(0)

# The sample data
sample_size = 90  # The sample size
successes = 55  # The number of successes in the sample
sample_proportion = successes / sample_size

# The null hypothesis is p = 0.5
# The alternative hypothesis is p != 0.5
p0 = 0.5  # The hypothesized population proportion under the null hypothesis

# Calculation of the standard error
standard_error = np.sqrt((p0 * (1 - p0))/sample_size)

# Calculation of the z-score
z_score = (sample_proportion - p0) / standard_error

# Calculation of the two-tailed p-value
two_tailed_p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

print("Sample Proportion:", sample_proportion)
print("Z-score:", z_score)
print("Two tailed p-value:", two_tailed_p_value)

# We'll plot the standard normal distribution.
x = np.linspace(-4, 4, 900)
y = stats.norm.pdf(x)

plt.plot(x, y, label='The Standard Normal Distribution', color='black')
plt.axvline(x=z_score, color='green', linestyle='--', label='z-scores')
plt.axvline(x=-z_score, color='green', linestyle='--')

# We'll shade the rejection regions.
x_fill_right = np.linspace(z_score, 4, 900)
y_fill_right = stats.norm.pdf(x_fill_right)
plt.fill_between(x_fill_right, 0, y_fill_right, color='red', label='The Rejection Region In The Right Tail')

x_fill_left = np.linspace(-4, -z_score, 900)
y_fill_left = stats.norm.pdf(x_fill_left)
plt.fill_between(x_fill_left, 0, y_fill_left, color='red', label='The Rejection Region In The Left Tail')

plt.title('Two-Tailed Z-Test For Proportion')
plt.xlabel('Z-Score')
plt.ylabel('The Density')
plt.legend()
plt.show()
```

Confidence intervals form an important part of hypothesis testing. This is because the alpha($\alpha$) level can be determined from the given confidence interval. $\alpha$ is the significance level which is used to compute the confidence level. The confidence level is $100\cdot(1 - \alpha)\%$, i.e. the $\alpha$ of 5% indicates the 95% confidence level.

If we obtain the alpha value for a two tailed hypothesis testing, then we will divide this value by 2, which is $\frac{\alpha}{2}$. Also, the confidence level for two tailed hypothesis testing becomes $100\cdot(1 - \frac{\alpha}{2})\%$.

The confidence level is the overall capture rate if the method is used many times. The sample mean can vary from sample to sample. A% of the intervals capture the population mean . That is, the actual mean is located within the interval A% of the time.

Confidence interval = sample mean  margin of error

To find the confidence interval, the margin of error has to be known. The margin of error is dependent on the degree of confidence. The degree of confidence can change between 90% and 99.9%, but it is up to us to decide.

$
\text{Margin of Error} = z^* \cdot \frac{\sigma}{\sqrt{n}}$, where $\sigma$ is the population standard deviation, n is the size of the sample, and $z^*$ is the level of confidence.

If we combine the formulas for the confidence interval and margin of error, then we will obtain the equation

Confidence interval = sample mean  $z^* \cdot \frac{\sigma}{\sqrt{n}}$

Confidence interval = (sample mean - margin of error, sample mean + margin of error).

The example about this is in the following:

```python
import numpy as np
import scipy.stats as stats

data = np.array([20, 16, 7, 15, 10, 12, 2, 6, 7, 8])  # data is the array of the sample data points.

mean = np.mean(data)  # the sample mean of the data

std_dev = np.std(data, ddof=1)  # the sample standard deviation with Bessel's correction ddof=1.

n = len(data)  # the sample size of the data

confidence_level = 0.94   # the confidence level

# Computation of the critical value
alpha = 1 - confidence_level
z_critical = stats.norm.ppf(1 - alpha/2)

# Calculation of the margin of error
margin_of_error = z_critical * (std_dev / np.sqrt(n))

confidence_interval = (mean - margin_of_error, mean + margin_of_error)  # the confidence interval

print(f"Sample mean: {mean:.2f}")
print(f"Confidence interval: ({confidence_interval[0]:.2f}, {confidence_interval[1]:.2f})")
```

# Hypothesis Testing For T Test

If n<30, a small size, then we use s instead of $\sigma$, and t test can be implemented here.

For one sample, $$t = \frac{\bar{x} - \mu}{\frac{s}{\sqrt{n}}}$$, where $\bar{x}$ is the sample mean, $\mu$ is the hypothesized population mean, s is the standard deviation of the sample, and n is the sample size.

For two samples, $$t = \frac{\bar{x}_1 - \bar{x}_2}{\sqrt{\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}}}$$, where 
$\bar{x}_1$ and $\bar{x}_2$ are the sample means, and $s_1^2$ and $s_2^2$ are the sample variances, and
$n_1$ and $n_2$ are the sample sizes.


```python
import numpy as np
import scipy.stats as stats

# Generation of the sample data
np.random.seed(0)  # It sets the random number generator to the fixed state so that the results are reproducible.
# Let's take the mean = 4, the standard deviation = 5 and the sample size = 50.
sample_data = np.random.normal(loc = 4, scale = 5, size = 50) 

# Calculation of the mean and the standard deviation
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)  # Use ddof=1 for sample standard deviation

# We can set the confidence level.
confidence_level = 0.95

# Calculation of the confidence interval
n = len(sample_data)  # The size of the sample data
t = stats.t.ppf((1 + confidence_level) / 2, df=n-1)  # t-critical value

margin_of_error = t * sample_std / np.sqrt(n)  # Computing the margin of error for the confidence interval

# The calculation of the confidence interval
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print("The Sample Mean:", sample_mean)
print("The Standard Deviation:", sample_std)
print("The Confidence Interval:", confidence_interval)
```

The degrees of freedom are the maximum number of logically independent values, which can vary in the data sample. The degrees of freedom are calculated by subtracting one from the number of items in the sample of the data.

# Hypothesis Testing For Critical Region

All sets of values that lead to rejecting the null hypothesis lie in the critical region. Also, the value that seperates the critical region from the non-critical region is the critical value. Usually, one-sided tests have one critical value and two-sided test have two critical values. The critical values are determined so that the probability that the test statistic has a value in the rejection region of the test when the null hypothesis, which is true, equals the significance level that is denoted as $\alpha$.

The critical value of z-score is used when the sampling distribution is normal. Z-scores are used when the population standard deviation is known or when we have larger sample sizes. While the z-score can also be used to calculate the probability for standard deviations and small samples, we use the t distribution to calculate the probabilities in real life since we do not know the population variance.

Let's look at the examples of this.

```python
import numpy as np
import scipy.stats as stats

# The data and parameters of the sample
sample_mean = 10.5   # the sample mean
null_mean = 10   # the null hypothesis mean
sample_std = 2  # the standard deviation
sample_size = 40   # the sample size

alpha = 0.05  # 5% significance level 

# Calculation of the degrees of freedom
df = sample_size - 1

# stats.t.ppf function enable us to calculate the critical values for the two-tailed test.
t_critical_left = stats.t.ppf(alpha / 2, df)
t_critical_right = stats.t.ppf(1 - alpha / 2, df)

# Calculation of the test statistic
t_statistic = (sample_mean - null_mean)/(sample_std / np.sqrt(sample_size))

print("The Critical Region For Two-Tailed Test):")
print("t < {:.4f} or t > {:.4f}".format(t_critical_left, t_critical_right))

# We have to determine whether the test statistic falls in the critical region or not.
if t_statistic < t_critical_left or t_statistic > t_critical_right:
    print("Reject the null hypothesis")
else:
    print("Do not reject the null hypothesis")
```
