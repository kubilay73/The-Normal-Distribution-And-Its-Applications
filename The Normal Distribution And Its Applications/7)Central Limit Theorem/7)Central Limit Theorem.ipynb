# Central Limit Theorem

It is one of the most important results in probability theory. It states that the distribution of the sample variable approximates the normal distribution because the sample size becomes larger, regardless of the shape of population's actual distribution. In this case, given a sufficiently large sample size from a population with a finite level of variance, the mean of all sampled variables from the same population will be approximately equal to the mean of the whole population. In addition, according to the law of large numbers, the variances are approximately equal to the variance of the population because the sample size becomes larger.

Even though it was first improved by Abraham de Moivre in 1733, it was not formalized until 1920. Then, Hungarian mathematician George P√≥lya dubbed it the theorem.

There is an example in the following:

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# The seaborn library facilitates the process of forming complex visualizations and supplies better aesthetics out of the box which is compared to matplotlib.
# The parameters
num_samples = 400  # The number of the samples
sample_size = 40  # The size of the samples

# Generation of the data of uniform distribution
# 'size = (num_samples, sample_size)' specifies the shape of the output array.
uniform_data = np.random.uniform(low = 0, high = 1, size = (num_samples, sample_size))

# We have to compute the means of the samples.
# 'axis=1' specifies that the mean can be calculated along the rows. 
sample_means = np.mean(uniform_data, axis=1)

# We'll plot the distribution of the sample means.
plt.figure(figsize=(9, 6))  # Creation of the figure of size 9x6 inches
# The function sns.histplot from the Seaborn library creates the histogram plot that shows the frequency distribution of the dataset.
# 'kde=True' adds the Kernel Density Estimate to the plot, which is the smoothed curve that shows the distribution of the data.
# In 'bins=26', the range of sample means is divided into 26 bins.
# 'sample_means' is the 1D array of sample means that are calculated earlier.
sns.histplot(sample_means, bins=26, kde=True, color='red')

# The title and labels are added.
plt.title('The Distribution Of Sample Means')
plt.xlabel('The Sample Mean')
plt.ylabel('The Frequency')

plt.show()
```

In this case, the plot is an example of the distribution of the sample means of the uniform distribution. Even though the original data is uniformly distributed, the distribution of the sample means appears approximately normal. It shows the Central Limit Theorem, which explains that the means of sufficiently large samples from any distribution can tend to be normally distributed.

Since we can estimate and make conclusions about the population proportion, we need to know the sampling distribution of the sample proportion because the sample proportion is used to estimate the population proportion. 

Assume that we have a random sample of size n which is got from the population where some individuals have the certain characteristic, while some does not. Let x be the number of individuals with the certain characteristic in the sample. Then, $\hat{p} = \frac{x}{n}$ is the sample proportion and estimates p, which is the proportion of the population.

In addition, the expected value of pÃÇ is $E(\hat{p}) = \mu_{\hat{p}} = p$, and the variance of pÃÇ is $V(\hat{p}) = \frac{p(1-p)}{n}$.

Then, the standard deviation of the sampling distribution of pÃÇ is $\sigma_{\hat{p}} = \sqrt{\frac{p(1-p)}{n}}$ and the z-score is z = $\frac{\hat{p} - p}{\sqrt{\frac{p(1-p)}{n}}}$.

Since the sampling distribution of the sample proportion is approximately normal under certain conditions, we can use the normal approximation to find the probabilities; as a result, we have to convert pÃÇ to z-score. Then, in this case, we will have $$z = \frac{\hat{p} - \mu_{\hat{p}}}{\sigma}$$

For example, the last research shows that 78% of teenagers, whose ages are between 12 and 18, have access to smartphones. A
random sample of 198 teenagers is selected. What is the sampling distribution of the proportion of them have access to smartphones?

```python
import numpy as np
import matplotlib.pyplot as plt

true_proportion = 0.78  # True proportion from the population
sample_size = 198  # The sample size
num_samples = 8700  # The number of samples to draw 

# The sampling distribution is simulated.
sampling_distribution = np.random.binomial(n=sample_size, p=true_proportion, size=num_samples) / sample_size

# We'll simulate the sampling distribution.
plt.hist(sampling_distribution, bins=35, density=True, color='orange', edgecolor='purple')

plt.title('The Sampling Distribution Of Proportion Of Teenagers With Smartphone Access')

plt.xlabel('The Proportion')
plt.ylabel('The Density')

plt.grid(True)

plt.show()
```

If we find the probability that more than 73% of the sampled teenagers have access to smartphones, then we get

```python
probability_more_than_73 = np.mean(sampling_distribution > 0.73)

print("The probability that more than 68% of sampled teenagers have access to smartphones:", probability_more_than_73)
```

If we find the probability that less than 80% of the sampled teenagers have access to smartphones, then we have

```python
# Calculation of the probability
probability_less_than_80 = np.mean(sampling_distribution < 0.80)

print("The probability that less than 76% of sampled teenagers have access to smartphones:", probability_less_than_80)
```
