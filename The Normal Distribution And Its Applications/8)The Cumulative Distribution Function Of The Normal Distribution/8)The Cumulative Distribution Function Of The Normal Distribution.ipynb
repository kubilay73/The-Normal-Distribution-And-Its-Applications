# The Cumulative Distribution Function Of The Normal Distribution

If we find the general formula of its cumulative distribution function of the standard normal distribution, we have to integrate the probability density function. 

Then, the formula of the cumulative distribution function of the normal distribution is given by $$
\Phi(x) = P(Z \leq x) = \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{x} e^{-\frac{u^2}{2}} \, du
$$

The example of it is in the following:

```python
import numpy as np
from scipy.integrate import quad
# quad from scipy.integrate is imported to perform numerical integration.
# standard_normal_cdf(x) is defined to calculate the cumulative density function of the standard normal distribution at the value x.
def standard_normal_cdf(x):
# The lambda function integrand is defined to represent the integrand of the standard normal cumulative density function.
    integrand = lambda u: np.exp(-u**2 / 2) / np.sqrt(2 * np.pi)
# The function quad computes the definite integral of the integrand function in terms of u over the interval −∞ to x. 
# The function quad returns two values that are the result of the integration and the estimate of the absolute error in the result, which is _.
    cdf, _ = quad(integrand, -np.inf, x)
    
    return cdf
```

```python
# For example, take x = 1.6.
x = 1.6

cdf_value = standard_normal_cdf(x)

print("Φ(x) =", cdf_value)
```

If we draw the cumulative distribution function of the standard normal distribution, then we can obtain this.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# The values between -4 and 4 represent the range of x for which we want to calculate Φ(x).
x = np.linspace(-4, 4)

# Calculation of the cumulative distribution function of the standard normal distribution for the x values
phi_x = norm.cdf(x)

# We'll plot the function.
plt.plot(x, phi_x, label='Φ(x)')
plt.title('The Cumulative Distribution Function Of The Standard Normal Distribution')

plt.xlabel('x')

plt.ylabel('Φ(x)')

plt.grid(True)

plt.legend()
plt.show()
```

We'll look at the example of the cumulative distribution function of the standard normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# The values between -4 and 4 represent the range of x for which we want to calculate Φ(x).
x = np.linspace(-4, 4)

# Calculation of the cumulative distribution function of the standard normal distribution for the x values.
phi_x = norm.cdf(x)

# We'll plot the function.
plt.plot(x, phi_x, label='Φ(x)')

# We'll shade the area under the curve for x values between -infinity and 0.7.
shade_x = np.linspace(-4, 0.7)

plt.fill_between(shade_x, norm.cdf(shade_x), color='purple')

# Computation of the area
the_area = norm.cdf(0.7) - norm.cdf(-4)

print("The area under the curve between -infinity and 0.7 is:", the_area)

plt.title('The Cumulative Distribution Function Of The Standard Normal Distribution')

plt.xlabel('x')
plt.ylabel('Φ(x)')

plt.grid(True)

plt.legend()

plt.show()
```

If X $\sim N(μ, σ^2)$, then $Z = {\frac{X - \mu}{\sigma}}$, which is the random variable, is a standard normal random variable. That is, Z $\sim N(0,1)$.

To find the cumulative distribution function of X $\sim N(μ, σ^2)$, we can write;

$$F_X(x) = P(X \leq x) = P(\sigma Z + \mu \leq x) = P\left(Z \leq \frac{x - \mu}{\sigma}\right) = \Phi\left(\frac{x - \mu}{\sigma}\right)$$

If X is a normal random variable with mean μ and variance $σ^2$, i.e. X $\sim N(μ, σ^2)$, then

$$f_X(x) = \frac{1}{\sigma \sqrt{2\pi}} \exp\left(-\frac{(x - \mu)^2}{2\sigma^2}\right)$$, which is the density function of the normal distribution.

$F_X(x) = P(X \leq x) = \Phi\left(\frac{x - \mu}{\sigma}\right)$

$P(a < X \leq b) = \Phi\left(\frac{b - \mu}{\sigma}\right) - \Phi\left(\frac{a - \mu}{\sigma}\right)$

Let's look at the example.

```python
import numpy as np
from scipy.stats import norm

# Calculation of the cumulative density function values at a and b
Phi_a = norm.cdf(a)
Phi_b = norm.cdf(b)

# We'll calculate the probability P(a < X <= b).
the_probability = Phi_b - Phi_a

# For example, let's take a = -0.87 and b = 0.94.
a = -0.87
b = 0.94

print(f"P({a} < X <= {b}) = {the_probability:.4f}")
```
